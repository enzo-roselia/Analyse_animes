```python
import pandas as pd
df = pd.read_csv("dataanime.csv")
df.dtypes
```

```python
#supprimer colonnes + vÃ©rification
df.drop(["Description","Episodes","Status","Producers","Starting season","Broadcast time","Licensors","Sources","Duration"],axis=1,inplace=True)
df.dtypes
```

```python
import numpy as np
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
```

```python
#valeurs manquantes ?
missing_data = df.isnull()
missing_data.head(30)
#on a l'impression que non au vu des 30 premiers individus
```

```python
#VÃ©rification de l'observation
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
#AperÃ§u des moyennes des variables quantitatives
avg_score = df["Score"].astype("float64").mean(axis=0)
print("Average of score:", avg_score)

avg_score = df["Scored by"].astype("float64").mean(axis=0)
print("Average of scored by:", avg_score)

avg_score = df["Members"].astype("float64").mean(axis=0)
print("Average of members:", avg_score)

avg_score = df["Favorites"].astype("float64").mean(axis=0)
print("Average of favorites:", avg_score)
```

```python
#changer de type (passer d'objet Ã  string pour faciliter la manipulation) puis vÃ©rification
df[["Title", "Studios","Genres","Type","Rating","Start airing"]] = df[["Title", "Studios","Genres","Type","Rating","Start airing"]].astype("string")
df.replace("-", np.nan, inplace = True)
df.dropna(axis=0,inplace=True)
df["Start airing"]=df["Start airing"].str[:4]
df["Start airing"]=df["Start airing"].astype("int64")
df.dtypes
```

```python
avg_score = df["Start airing"].astype("int64").mean(axis=0)
print("Average of start airing:", avg_score)
```

```python
#Histogramme des notes pour mieux visualiser la rÃ©partition
df["Score"]=df["Score"].astype(float, copy=True)
plt.pyplot.hist(df["Score"])
#titres des axes
plt.pyplot.xlabel("Score")
plt.pyplot.ylabel("FrÃ©quence")
plt.pyplot.title("Histogramme des scores")
```

```python
#Je rajoute une variable qualitative selon le score pour faciliter certaines techniques d'analyse
bins = np.linspace(min(df["Score"]), max(df["Score"]), 4)
bins
```

```python
#grouper par score
group_names = ['Mediocre', 'Bon', 'Excellent']
df['Niveau'] = pd.cut(df['Score'], bins, labels=group_names, include_lowest=True )
df["Niveau"].value_counts()
```

```python
#Histogramme de la nouvelle variable
pyplot.bar(group_names, df["Niveau"].value_counts())

plt.pyplot.xlabel("Niveau")
plt.pyplot.ylabel("FrÃ©quence")
plt.pyplot.title("Histogramme des niveaux")
```

```python
#Je vÃ©rifie que la nouvelle variable est enregistrÃ©e
df.columns
```

```python
#Nouvelle variable, je change les types en dummy car trÃ¨s peu de modalitÃ©s (6 diffÃ©rentes)
dummy_variable_1 = pd.get_dummies(df["Type"])
dummy_variable_1.head()
```

```python
#Je rajoute les dummies au dataframe
df = pd.concat([df, dummy_variable_1], axis=1)
```

```python
#verification des changements
df.dtypes
```

```python
#Je sauvegarde les changements
df["Niveau"]=df["Niveau"].astype("string")
df.to_csv('clean_df.csv',index=False)
```
